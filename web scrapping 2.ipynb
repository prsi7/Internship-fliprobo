# Q6 : Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes: 1. Brand 2. Product Description 3. Price 4. Discount %

```python
driver=webdriver.Chrome(r'C:\Users\Prsi\Desktop\ML\chromedriver.exe')
```

```python
def fetch_data():
    # get web element for brand name
    brand_wbe=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
    #get brand name text
    for i in brand_wbe:
        brands.append(i.text)
        if len(brands)==100:
            break
        
    # get web element for Price
    price_wbe=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
    # get price text from above
    for i in price_wbe:
        prices.append(i.text)
        if len(prices)==100:
            break

    #get web element for discount
    disc_wbe=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")
    # get text from above
    for i in disc_wbe:
        discounts.append(i.text)
        if len(discounts)==100:
            break

    #get web element for product description
    descr_wbe=driver.find_elements_by_xpath("//a[@class='IRpwTa']")
    # get text from above
    for i in descr_wbe:
        descriptions.append(i.text)
        if len(descriptions)==100:
            break
```

```python
# enter the url
url="https://www.flipkart.com/"
driver.get(url)

#add timer sleep
time.sleep(2)

# get web element for login email search bar
email_wbe=driver.find_element_by_xpath("//input[contains(@class,'_2IX_2-')]")#@class='_2IX_2- _2LYh3d VJZDxU'
# enter the email id
email_wbe.send_keys("8208507760")

#get web element for password bar
pwd_wbe=driver.find_element_by_xpath("//input[@type='password']")#class='_2IX_2- _3mctLh VJZDxU'
# enter the password
pwd_wbe.send_keys("Newpassword@72")

# get the webelement for login button
login_btn=driver.find_element_by_xpath("//button[@class='_2KpZ6l _2HKlqd _3AWRsL']")
#click on login button
login_btn.click()

#add timer sleep
time.sleep(3)

# get the web element for search bar
search_wbe=driver.find_element_by_xpath("//input[@name='q' and @placeholder='Search for products, brands and more']")
# write into search bar and pressing ENTER 
search_wbe.send_keys("sunglasses",Keys.ENTER)

#add timer sleep
time.sleep(3)


brands=[]
brands.clear()
        
prices=[]
prices.clear()

discounts=[]
discounts.clear()

descriptions=[]
descriptions.clear()

# get web element for page button
page_wbe=driver.find_elements_by_xpath("//a[contains(@class,'ge-49M')]")
# get urls of each page
page_urls=[]
page_urls.clear()

#to store page numbers
page_numbers=[]
page_numbers.clear()

# get urls of each next page from webelement
for i in page_wbe:
    temp_url= i.get_attribute('href') # getting the page urls
    page_urls.append(temp_url)        # storing the urls in a list
    page_numbers.append(i.text)
    
# counter for iteration    
counter=0   
# iterate over the webpages
for url in page_urls:
    if counter==0: # if 1st page, 
        fetch_data()
        counter=1 # change counter value to go to next webpage in next iteration
    else:
        if len(brands)==100: # checking if 100 items are scraped till this point during iterations
            break
        else:
            # store main window handle in a variable
            main_window=driver.current_window_handle
            
            # create a new tab to open the url
            driver.execute_script("window.open('about:blank','secondtab');")
            
            # switch driver to new tab
            driver.switch_to.window('secondtab')
            
            # open the new url in new tab
            driver.get(url)
            
            # give time to load webpage
            time.sleep(3)
            
            # fetch data from current page
            fetch_data()
            
            # close the new tab
            driver.close()
            
            # switch driver to main page
            driver.switch_to.window(main_window)


#close the driver
driver.quit()

# create the dataframe
glasses=pd.DataFrame({})
glasses['Brand Name']=brands
glasses['Price']=prices
glasses['Discount']=discounts
glasses['Product Description']=descriptions
```

```python
def fetch_data():
    # get webelement for brand name
    brand_wbe=driver.find_elements_by_xpath("//div[@class='product-productMetaInfo']/h3")
    #get text
    for i in brand_wbe:
        brands.append(i.text)
        if len(brands)==100:
            break
    
        
    #get webelement for product description
    descr_wbe=driver.find_elements_by_xpath("//div[@class='product-productMetaInfo']/h4[@class='product-product']")
    # get text
    for i in descr_wbe:
        descriptions.append(i.text)
        if len(descriptions)==100:
            break
    
    #get webelement for price
    price_wbe=driver.find_elements_by_xpath("//div[@class='product-price']/span[1]")
    #get text
    for i in price_wbe:
        if len(prices)==100:
            break
        else:
            temp_text=i.text
            if len(temp_text)<9:
                prices.append(temp_text)
            else:
                temp_text2=temp_text.split('Rs. ')[1]
                prices.append("Rs. "+temp_text2)
```

```python
start_page=0
end_page=0
for page in range(start_page,end_page+1): 
    nxt_button=driver.find_element_by_xpath("//li[@class='pagination-next']")
    
    #creating empty lists
    shoe_names=[]
    s_desc=[]
    short_desc=[]
    price=[]
    
    #for scrapping shoe brand names
    Names=driver.find_elements_by_xpath("//div[@class='product-productMetaInfo']/h3")
    for i in Names:
        shoe_names.append(i.text)
    
    #for scrapping shoe short-description
    desc=driver.find_elements_by_xpath("//div[@class='product-productMetaInfo']/h4")
    for i in desc:
        s_desc.append(i.text)
    #As, the s_desc list contain blank description in every alternate index, so removing the blank or null description
    for j in range(0,len(s_desc),2):
        short_desc.append(s_desc[j])
    
    #for scrapping shoe prices
    desc=driver.find_elements_by_xpath("//div[@class='product-price']")
    for i in desc:
        price.append(i.text)
    
    #for scrapping the datas from next pages.
    if nxt_button.text=='next':
            nxt_button.click()
            time.sleep(5)
    
    # create the dataframe
shoes=pd.DataFrame({})
shoes['Brand Name']=shoe_names
shoes['Price']=price
shoes['Product Description']=short_desc 

shoes
```

# Q10: Go to webpage https://www.amazon.in/Enter “Laptop” in the search field and then click the search icon.Then set CPU Type filter to “Intel Core i7” and “Intel Core i9” After setting the filters scrape first 10 laptops data.You have to scrape 3 attributes for each laptop:

```python
driver=webdriver.Chrome(r'C:\Users\Prsi\Desktop\ML\chromedriver.exe')
```

```python
# enter the url
url='https://www.amazon.in/'
driver.get(url)

#add timer
time.sleep(3)

driver.maximize_window()

# get the webelement for search bar
search_wbe=driver.find_element_by_xpath("//input[@type='text' and @id='twotabsearchtextbox']")
#enter into search bar
search_wbe.send_keys("Laptop",Keys.ENTER)


#add timer
time.sleep(3)

# get webelement of 1st filter
i7_wbe=driver.find_element_by_xpath("//li[@id='p_n_feature_thirteen_browse-bin/12598163031']//following::i")
#click
i7_wbe.click()

# get webelement for title
title_wbe=driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")
titles=[]
titles.clear()
for i in title_wbe:
    titles.append(i.text)
    if len(titles)==10:
        break
        
        
ratings=[]
ratings.clear()
prices=[]
prices.clear()

for wbe in title_wbe:    
    #click on the title
    wbe.click()
    
    #add timer
    time.sleep(3)

    #swtich to new window
    driver.switch_to.window(driver.window_handles[1])
    
    #add timer
    time.sleep(3)
    
    # get webelement for rating
    try:
        rating_wbe=driver.find_element_by_xpath("//div[@class='a-section a-spacing-none a-spacing-top-mini cr-widget-ACR']")
        # get rating text
        if rating_wbe.text=='':
            ratings.append("Not Available")
        else:
            ratings.append(rating_wbe.text.replace("\n",",").split(",")[0])
    except NoSuchElementException:
        ratings.append("Not Available")
        
    #add timer
    time.sleep(3)
    
    # get webelement for price
    try:
        price_wbe=driver.find_element_by_xpath("//span[@id='priceblock_ourprice']")
        #get price text
        prices.append(price_wbe.text)
    except NoSuchElementException:
        try:
            price_wbe=driver.find_element_by_xpath("//span[@id='priceblock_dealprice']")
            #get price text
            prices.append(price_wbe.text+"(Deal Price)")
        except NoSuchElementException:
            prices.append("Not Available")
    
    
    # close the new window
    driver.close()
    
    #switch to main window
    driver.switch_to.window(driver.window_handles[0])
    
    if len(prices)==10:
        break

        
# create dataframes
lp_models=pd.DataFrame({})
lp_models['CPU type']=pd.Series(['Intel core i7' for x in range(len(titles))])
lp_models['Model']=titles
lp_models['Rating']=ratings
lp_models['Price']=prices


#get webelement for clear filter
clear_wbe=driver.find_element_by_xpath("//li[@class='a-spacing-micro s-list-item']//following::a")
# clear the 1st filter
clear_wbe.click()
#add timer
time.sleep(3)

# get webelement for 2nd filter
i9_wbe=driver.find_element_by_xpath("//li[@id='p_n_feature_thirteen_browse-bin/16757432031']//following::i")
#click
i9_wbe.click()

#add timer
time.sleep(3)

# get webelement for title
title_wbe_1=driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")
titles_1=[]
titles_1.clear()
for i in title_wbe_1:
    titles_1.append(i.text)
    if len(titles_1)==10:
        break
        
        
ratings_1=[]
ratings_1.clear()
prices_1=[]
prices_1.clear()

for wbe in title_wbe_1:    
    #click on the title
    wbe.click()
    
    #add timer
    time.sleep(3)

    #swtich to new window
    driver.switch_to.window(driver.window_handles[1])
    
    #add timer
    time.sleep(3)
    
    # get webelement for rating
    try:
        rating_wbe_1=driver.find_element_by_xpath("//div[@class='a-section a-spacing-none a-spacing-top-mini cr-widget-ACR']")
        # get rating text
        if (rating_wbe_1.text=='' or rating_wbe_1.text=='None'):
            ratings_1.append("Not Available")
        else:
            ratings_1.append(rating_wbe_1.text.replace("\n",",").split(",")[0])
    except NoSuchElementException:
        ratings_1.append("Not Available")
        
    #add timer
    time.sleep(3)
    
    # get webelement for price
    try:
        price_wbe_1=driver.find_element_by_xpath("//span[@id='priceblock_ourprice']")
        #get price text
        prices_1.append(price_wbe_1.text)
    except NoSuchElementException:
        try:
            price_wbe_1=driver.find_element_by_xpath("//span[@id='priceblock_dealprice']")
            #get price text
            prices_1.append(price_wbe_1.text+"(Deal Price)")
        except NoSuchElementException:
            prices_1.append("Not Available")
    
    
    # close the new window
    driver.close()
    
    #switch to main window
    driver.switch_to.window(driver.window_handles[0])
    
    if len(prices_1)==10:
        break


# create dataframes
lp_models_1=pd.DataFrame({})
lp_models_1['CPU type']=pd.Series(['Intel core i9' for x in range(len(titles_1))])
lp_models_1['Model']=titles_1
lp_models_1['Rating']=ratings_1
lp_models_1['Price']=prices_1

lp_models.append(lp_models_1,ignore_index=True)
```
